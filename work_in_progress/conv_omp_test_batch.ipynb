# Dictionary definition

```python
import numpy as np
from csc.atoms import *
from csc.dictionary import *
import matplotlib.pyplot as plt

# atom parameters
bmin = 0.005
bmax = 0.60
b_step = 0.005
b_values = np.arange(bmin, bmax, b_step)

ymin = 0.01
ymax = 0.034
y_step = 0.002
y_values = np.arange(ymin, ymax, y_step)

s_values = [1]

dictionary = ZSDictionary.from_values(b_values, y_values, s_values)
print('The dictionary has {} different atoms'.format(len(dictionary)))
```

# OMP test batch

## ``sparsity_level`` = 1

```python
signals1, reconstructions1, infos1 = dictionary.ompTestBatch(
    batch_size=10, 
    signals_length=2*dictionary.getAtomsLength(),
    sparsity_level=1,
    noise_level=0.0
    )
```

```python
fig, axs = plt.subplots(len(signals1), 1, figsize=(18, 3*len(signals1)), sharex=True)

for i in range(len(signals1)) :
    axs[i].plot(signals1[i], label='original')
    axs[i].plot(reconstructions1[i], label='recovered')
    original_atoms = infos1[i]['original']
    recovered_atoms = infos1[i]['recovered']
    matching_idx = infos1[i]['matching_idx']
    matching_err = infos1[i]['matching_err']
    ax_title = 'Matching errors : mse = {:.2e} |'.format(infos1[i]['mse'])
    for a_idx, err in matching_err.items() :
        ax_title += '|  A{} : {:.2f}%  |'.format(a_idx, err*100)
    ax_title += '|'
    axs[i].set_title(ax_title)
    axs[i].legend()
plt.show()
```
